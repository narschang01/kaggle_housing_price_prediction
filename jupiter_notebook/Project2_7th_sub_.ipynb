# Define Feature Matrix & Target

```python
features = ['1st_flr_sf', '2nd_flr_sf', 'garage_area', 'gr_liv_area',
            'open_porch_sf','overall_qual', 'total_bsmt_sf', 'year_remod/add', 'totrms_abvgrd', 
            'year_built', 'neighborhood_NoRidge', 'fbxgrla', 'frxgrla',
            'neighborhood_NridgHt', 'neighborhood_OldTown','neighborhood_StoneBr', 'neighborhood_NAmes', 
            'neighborhood_IDOTRR', 'neighborhood_Edwards', 'neighborhood_Sawyer', 'neighborhood_BrkSide' ]
```

```python
df[features].isnull().sum()
```

```python
y = df['saleprice']
X = df[features]
```

#### Train-Test-Split

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, random_state = 308)
```

## Standardscaler transform

```python
ss = StandardScaler()
```

```python
ss.fit(X_train)
```

```python
X_train_sc = ss.transform(X_train)
```

```python
X_valid_sc = ss.transform(X_valid)
```

#### Instantiate, Fit, Score Model

```python
lr = LinearRegression()
```

```python
lr.fit(X_train_sc, y_train)
```

```python
cross_val_score(lr, X_train_sc, y_train)
```

```python
cross_val_score(lr, X_valid_sc, y_valid).mean()
```

```python
lr.score(X_train_sc, y_train)
```

```python
cross_val_score(lr, X_valid_sc, y_valid)
```

```python
cross_val_score(lr, X_valid_sc, y_valid).mean()
```

```python
lr.score(X_valid_sc, y_valid)
```

```python
lr.score(X_train_sc, y_train), lr.score(X_valid_sc, y_valid)
```

```python
list(zip(X.columns, lr.coef_))
```
